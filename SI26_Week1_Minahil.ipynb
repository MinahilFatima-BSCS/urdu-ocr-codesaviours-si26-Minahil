**What is OCR (Optical Character Recognition)?**

OCR (Optical Character Recognition) is a technology that converts text from images, scanned documents, or photos into editable digital text. It helps computers recognize printed or handwritten characters automatically. OCR is widely used to digitize books, forms, invoices, and other paper documents, making them easier to search, edit, and store.

**Why is Urdu OCR harder than English OCR?**

Urdu OCR is more difficult because Urdu is written in a connected script where the shape of a letter changes depending on its position in a word. Many Urdu letters look very similar and can only be distinguished by the number or position of dots. Different fonts, writing styles, and image quality also make it more challenging for OCR systems to recognize Urdu text accurately.

**What are 2 real-world situations where Urdu OCR would be useful?**

One important use of Urdu OCR is digitizing old Urdu books, newspapers, and historical documents so they can be searched and preserved electronically. Another useful application is converting printed government forms, educational materials, or business documents into editable digital text, which saves time and reduces manual typing errors.

```python
# ===============================
# Generate Synthetic Urdu Images
# ===============================

# Install required packages
!apt-get -qq update
!apt-get -qq install fonts-noto-core
!pip -q install arabic-reshaper python-bidi

import os
from PIL import Image, ImageDraw, ImageFont
import arabic_reshaper
from bidi.algorithm import get_display

# Create folder
output_folder = "data/raw/synthetic"
os.makedirs(output_folder, exist_ok=True)

# Urdu Texts
urdu_texts = [
    "پاکستان ایک خوبصورت ملک ہے",
    "اردو ہماری قومی زبان ہے",
    "تعلیم ہر انسان کا بنیادی حق ہے",
    "ہم روزانہ نئی چیزیں سیکھتے ہیں",
    "کمپیوٹر سائنس ایک دلچسپ مضمون ہے",
    "مشقت کامیابی کی کنجی ہے",
    "وقت بہت قیمتی ہے",
    "ایمانداری بہترین پالیسی ہے",
    "مطالعہ علم میں اضافہ کرتا ہے",
    "صحت اللہ کی نعمت ہے",
    "دوستی ایک خوبصورت رشتہ ہے",
    "درخت ماحول کو صاف رکھتے ہیں",
    "ہمیں صفائی کا خیال رکھنا چاہیے",
    "محنت کبھی ضائع نہیں جاتی",
    "کتاب بہترین دوست ہے",
    "پاکستان کے چار موسم ہیں",
    "سچ ہمیشہ کامیاب ہوتا ہے",
    "علم روشنی ہے",
    "ہر دن ایک نئی امید لاتا ہے",
    "زندگی ایک خوبصورت سفر ہے"
]

# Font
font_path = "/usr/share/fonts/truetype/noto/NotoNaskhArabic-Regular.ttf"
font = ImageFont.truetype(font_path, 42)

# Generate Images
for i, text in enumerate(urdu_texts):

    reshaped = arabic_reshaper.reshape(text)
    bidi_text = get_display(reshaped)

    img = Image.new("RGB", (1000, 150), "white")
    draw = ImageDraw.Draw(img)

    bbox = draw.textbbox((0,0), bidi_text, font=font)
    text_width = bbox[2]-bbox[0]
    text_height = bbox[3]-bbox[1]

    x = (1000-text_width)//2
    y = (150-text_height)//2

    draw.text((x,y), bidi_text, fill="black", font=font)

    img.save(f"{output_folder}/synthetic_{i+1}.png")

print(f"✅ {len(urdu_texts)} Synthetic Images Generated Successfully!")
print("Saved in:", output_folder)
```
